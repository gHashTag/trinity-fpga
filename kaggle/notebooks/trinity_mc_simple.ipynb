No kbench dependency - pure evaluation

```python
import pandas as pd
import kaggle
import glob

print("✅ Imports successful")
```

```python
# Download dataset
!mkdir -p /kaggle/working/datasets
kaggle.api.dataset_download_files(
    'playra/trinity-cognitive-probes-thlp-mc',
    path='/kaggle/working/datasets/',
    unzip=True
)
```

```python
# Load MC data
csv_files = glob.glob('/kaggle/working/datasets/**/*.csv', recursive=True)
csv_path = [f for f in csv_files if 'thlp_mc.csv' in f][0]

df = pd.read_csv(csv_path)
mc_df = df[df['question_type'] == 'mc'].copy()

# Take 50 items for test
eval_df = mc_df.head(50)

print(f"📊 Loaded {len(eval_df)} MC questions from {len(df)} total rows")
print(f"\nSample question:")
print(f"Q: {eval_df.iloc[0]['question'][:100]}...")
print(f"Choices: {eval_df.iloc[0]['choices'][:100]}...")
print(f"Answer: {eval_df.iloc[0]['answer']}")
```

```python
# Mock evaluation (random baseline)
import random
random.seed(42)

def match_mc_answer(response: str, expected: str) -> bool:
    """MC format: compare first letter only."""
    if not response or not expected:
        return False
    got = response.strip().upper()[0]
    exp = str(expected).strip().upper()[0]
    return got == exp

correct = 0
for _, row in eval_df.iterrows():
    # Random baseline
    mock_answer = random.choice(['A', 'B', 'C', 'D'])
    if match_mc_answer(mock_answer, row['answer']):
        correct += 1

accuracy = correct / len(eval_df)
print(f"\n🏆 Random Baseline: {accuracy:.1%} ({correct}/{len(eval_df)})")
print(f"\nExpected: ~25% for random, ~100% for model")
```
